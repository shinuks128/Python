```python
import numpy as np
import pandas as pd

# Exercise 1: Create array and reshape to 2x5 matrix
arr1 = np.arange(1, 11).reshape(2, 5)
print("Exercise 1:\n", arr1)

# Exercise 2: Extract elements between 5th and 15th index (inclusive of 5, exclusive of 15)
arr2 = np.arange(1, 21)
subset = arr2[5:15]
print("\nExercise 2:\n", subset)

# Exercise 3: Create series and add new item
fruits = pd.Series({'apples': 3, 'bananas': 2, 'oranges': 1})
fruits['pears'] = 4
print("\nExercise 3:\n", fruits)

# Exercise 4: Create dataframe with 10 rows
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Hannah', 'Ian', 'Jane'],
    'age': [25, 32, 28, 45, 30, 29, 35, 22, 40, 27],
    'gender': ['F', 'M', 'M', 'M', 'F', 'M', 'F', 'F', 'M', 'F']
}
df = pd.DataFrame(data)
print("\nExercise 4:\n", df)

# Exercise 5: Add new column 'occupation'
df['occupation'] = ['Programmer', 'Manager', 'Analyst', 'Programmer', 'Manager', 
                    'Analyst', 'Programmer', 'Manager', 'Analyst', 'Programmer']
print("\nExercise 5:\n", df)

# Exercise 6: Select rows where age >= 30
age_filter = df[df['age'] >= 30]
print("\nExercise 6:\n", age_filter)

# Exercise 7: Save to CSV, read it back, and display
df.to_csv('people.csv', index=False)
df_from_csv = pd.read_csv('people.csv')
print("\nExercise 7:\n", df_from_csv)
```
